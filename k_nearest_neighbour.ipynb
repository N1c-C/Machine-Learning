# **Nearest Neighbour Classification**
An example of a nearest neigbour classifier using the scikit learn datasets and functions.

```python
# Dependicies
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

# **Dataset**

The breast cancer dataset from `sklearn` is used for the example.
The dataset has two classes 'malignant'/'benign', 30 dimensions and 569 samples.

For this notebook the dataset is not loaded as a pandas dataframe and the dataset has the followinf attributtes.



*   data
*   target
* feature_names
* target_names




```python
from sklearn.datasets import load_breast_cancer
bc_data = load_breast_cancer()

# View attributes

print('Number of samples: ', len(bc_data.data))
print(bc_data.feature_names)
print(f'Row 0 values {bc_data.data[0]}. Class {bc_data.target_names[bc_data.target[0]]}')
```

Split Data - 80% Train and 20% Test

```python
x_train, x_test, labels_train, labels_test = train_test_split(
  bc_data.data,
  bc_data.target,
  test_size=0.2,
  random_state=42
)
```

Iteratively find the best accuarcy for the model with different numbers (k) of nearest neighbours to make the classification

```python
k_val=[]
score=[]
for k in range(1,101):
  classify = KNeighborsClassifier(k)
  classify.fit(x_train, labels_train)
  k_val.append(k)
  score.append(classify.score(x_test, labels_test))
```

Plot k-values verses accuarcy for the model



```python
plt.plot(k_val, score)
plt.title('k-Nearest Neighbour Classification - The effect of k on the overall accuracy')
plt.ylabel('Model accuracy')
plt.xlabel('k - Number of nearest neighbours taken into account')
plt.show()
```
